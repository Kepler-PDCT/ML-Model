```python
from augmentation import get_train_transforms, get_test_transforms

import torch
import torch.nn as nn
from torch.nn import functional as F

from torch.utils.data import DataLoader
from torchvision.datasets import ImageFolder
from torchvision import transforms
from torchvision.utils import make_grid

import matplotlib.pyplot as plt
import numpy as np
```

```python
train_transforms = get_train_transforms()
train_dataset = ImageFolder("./dataset/train", transform=train_transforms)
train_dataloader = DataLoader(train_dataset, batch_size=6*9, shuffle=True)


val_transforms = get_test_transforms()
val_dataset = ImageFolder("./dataset/val", transform=val_transforms)
val_dataloader = DataLoader(val_dataset, batch_size=6*9, shuffle=True)
```

```python
dataiter = iter(train_dataloader)
images, labels = dataiter.next()

img_grid = make_grid(images["image"], normalize=True, nrow=6)
img_grid = np.moveaxis(img_grid.numpy(), 0, 2)
plt.figure(figsize=(4*5,6*5))
plt.axis('off')
plt.imshow(img_grid, interpolation="spline36")
#plt.savefig("train.png", bbox_inches='tight')
```

```python
dataiter = iter(val_dataloader)
images, labels = dataiter.next()

labels = [[l] for l in labels]
#labels = labels.unsqueeze(1)
#print(labels.unsqueeze(1).shape)

#img_grid = make_grid_with_labels(images["image"], labels, normalize=True, nrow=4, limit=None)
img_grid = make_grid(images["image"], normalize=True, nrow=6)
img_grid = np.moveaxis(img_grid.numpy(), 0, 2)
plt.figure(figsize=(4*5,6*5))
plt.axis('off')
plt.imshow(img_grid, interpolation="spline36")
plt.savefig("test.png", bbox_inches='tight')
```
